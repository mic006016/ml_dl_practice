```python
def connect():
    CoffeeBean_URL = "https://www.coffeebeankorea.com/store/store.asp"
    wd = wb.Chrome()
    return wd
```

```python
def coffee_bean(no):
    wd.get(CoffeeBean_URL)
    shop = ''
    func = 'storePop2("' + str(no) + '")'
    try:
        time.sleep(1)
        wd.execute_script(func)
        time.sleep(1)
        html = wd.page_source
    #     print(html)
        soupCB = BeautifulSoup(html, 'html.parser')
        # #matizCoverLayer0Content > div > div > div.store_txt > h2
        store_name_h2 = soupCB.select("div.store_txt > h2")
        print('1---', store_name_h2)
        store_name = store_name_h2[0].string
        print('2---', store_name)  #매장 이름 출력하기
        store_info = soupCB.select("div.store_txt > table.store_table > tbody > tr > td")
        print('3---', store_info)
        store_address = store_info[2].getText()
        print('4---', store_address)
        store_phone = store_info[3].getText()
        print('5---', store_phone)
        #shop = store_name + ',' + store_address + ',' + store_phone
        shop = []
        shop.append(store_name)
        shop.append(store_address)
        shop.append(store_phone)
        print('크롤링 결과>>>', shop)
    except:
        print('해당번호없음')
    return shop
```

```python
result2 = []
wd = None
wd = connect()
for no in range(1, 20):
    print('=========' + str(no) + '===========')
    shop = coffee_bean(no)
    print(shop)
```

```python
def coffee_bean(no) :
    wd.get(CoffeeBean_URL)
    time.sleep(3)
    wd.execute_script("storePop2(" + str(no) + " )")
    time.sleep(3)
    
    html = wd.page_source #팝업창을 다운로드
    soupCB = BeautifulSoup(html, 'html.parser')
    store_name_h2 = soupCB.select("div.store_txt > h2")
    print("===> " , store_name_h2)
    # time.sleep(5)
    store_name = store_name_h2[0].string
    print(store_name)  #매장 이름 출력하기

    store_one = [] #store 1개에 대한 정보를 넣자.
    # [가게이름, 영업시간, 주차, 주소, 전화번호, 기타 정보]
    store_one.append(store_name)

    store_info = soupCB.select('#matizCoverLayer0Content > div > div > div.store_txt > table > tbody > tr > td')
    time.sleep(5)
    
    # 매장정보 가지고 와서 리스트에 다 모아주자!
    for i in range(0,4):
        print(store_info[i].text.strip())
        store_one.append(store_info[i].text.strip())
    print(store_one)
    
    return store_one
```

```python
result3 = []
wd = None
wd = connect()
for no in [3, 6, 12]:
    print('=========' + str(no) + '===========')
    try:
        shop = coffee_bean(no)
        print(shop)
        result3.append(shop)
    except:
        print("없음")
len(result3)
```

```python
df = pd.DataFrame(result3,
                  columns = ['name','time','car','addr','tel'])
df
```

```python
df.to_csv("coffeebean_store.csv", index=False, encoding="utf-8")
```
