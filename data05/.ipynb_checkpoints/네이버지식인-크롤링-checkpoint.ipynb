```python
# 한 페이지를 가지고 오는 것을 함수로 만들어버리자.!
def naver_kin_one(page):
    from bs4 import BeautifulSoup
    from urllib import request
    import pandas as pd
    import datetime

    con2 = request.urlopen('https://kin.naver.com/search/list.naver?query=%EC%B7%A8%EC%97%85&page=' + str(page))
    doc2 = BeautifulSoup(con2, 'html.parser')
    
    for i in range(1, 11): # 1~10
        title_list = doc2.select('#s_content > div.section > ul > li:nth-child(' + str(i) + ') > dl > dt > a')
        print(title_list[0].text.strip())
```

```python
naver_kin_one(1)
```

```python
naver_kin_one(2)
```

```python
naver_kin_one(100)
```

```python
# 여러 페이지를 가지고 오는 것은 함수를 여러번 실행하면 됨.!
for page in range(1, 201): # 1~200
    try:
        naver_kin_one(page)
        print('page 번호', page, '--------------------------------')
    except:
        break
```
