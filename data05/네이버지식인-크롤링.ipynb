```python
def naver_kin_one(page):
    from bs4 import BeautifulSoup
    from urllib import request
    import pandas as pd
    import datetime

    con2 = request.urlopen('https://kin.naver.com/search/list.naver?query=%EC%B7%A8%EC%97%85&page=' + str(page)) #005930
    doc2 = BeautifulSoup(con2, 'html.parser')

    rows_10 = [] # 한 페이지에 10개를 모을 리스트
    for i in range(1, 11): #1~10
        # 제목 추출
        title_list2 = doc2.select('#s_content > div.section > ul > li:nth-child(' + str(i) + ') > dl > dt > a') 
        print(title_list2[0].text.strip())
        title = title_list2[0].text.strip()

        # 날짜 추출
        date_list = doc2.select('#s_content > div.section > ul > li:nth-child(' + str(i) + ') > dl > dd.txt_inline') 
        print(date_list[0].text.strip())
        date = date_list[0].text.strip()
        
        # 내용 추출
        content_list = doc2.select('#s_content > div.section > ul > li:nth-child(' + str(i) + ') > dl > dd:nth-child(3)') 
        print(content_list[0].text.strip())
        content = content_list[0].text.strip()

        row = [title, date, content]
        rows_10.append(row)
    return rows_10
```

```python
naver_kin_one(1)
```

```python
for page in range(1, 201):
    try:
        print("page >> ", page, "------------------------------")
        result_1 = naver_kin_one(page)
        print(result_1)
    except:
        print("연결 끊김.!!!")
        break;    
```
