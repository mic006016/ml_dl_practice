```python
import requests # 파이써넹서 http통신을 할 수 있는 최근 라이브러리
```

```python
# 주소?query=자동차
# http(header+body)
# http - 검색(get, header에 요청한 주소 + 전달값)

# http header에는 보통 인증정보 + 받을 데이터 형식 지정
header_data= {
    'accept': 'application/json',
    'Authorization': 'VXgWV/f5wELzapZjJF7FEqqtFhPbNA8KBIWia3SXOCvY+CXEUok1DhybfbP4USE/g2b/QYT62C2mw3zcSWj2pg==',
}

param_data = (
    ('page', '1'),
    ('perPage', '10'),
    ('serviceKey', 'VXgWV/f5wELzapZjJF7FEqqtFhPbNA8KBIWia3SXOCvY+CXEUok1DhybfbP4USE/g2b/QYT62C2mw3zcSWj2pg=='),
)

url = 'https://api.odcloud.kr/api/15069309/v1/uddi:973ad0df-617a-4565-8e82-8ea7869e75d4'
## decode key가 적용되었음.
## api마다 연결이 되지 않는다고 한다면, encode, decode를 바꾸어 테스트해볼 것.
response = requests.get(url, headers=header_data, params=param_data)
response # response(http header+body, header(status, message))
```

```python
json_data = response.json()
# http의 body에 온 데이터를 json형태로 저장해라!
# python에서는 {키:값}형태가 딕셔너리(dict)형태로 저장함.
json_data
```

```python
json_data2 = response.text # http의 body에 있는 것 그대로 꺼내오겠음.
# json 형태의 str이 들어있음.
json_data2
```

```python
one_row = []
one_row.append(data_values[0]['제목']) # 제목
one_row.append(data_values[0]['본문']) # 본문
one_row.append(data_values[0]['통합 분류1']) # 통합 분류1
one_row.append(data_values[0]['일자']) # 일자
one_row.append(data_values[0]['주소']) # 주소
one_row.append(data_values[0]['언론사']) # 언론사

print(one_row)
```

```python
total_row = [] # 2차원리스트 1차원리스트를 모아놓은 것.
# 하나의 dict은 하나의 리스트임.
# 여러개는 [[row], [row], [row]] 2차원리스트

for no in range(10): # 0~9
    one_row = []
    one_row.append(data_values[no]['제목'])
    one_row.append(data_values[no]['본문'])
    one_row.append(data_values[no]['통합 분류1'])
    one_row.append(data_values[no]['일자'])
    one_row.append(data_values[no]['주소'])
    one_row.append(data_values[no]['언론사'])

    # print(one_row)
    
    total_row.append(one_row)
    
    print(len(total_row), '페이지')
    print(total_row[no])

len(total_row)
```
