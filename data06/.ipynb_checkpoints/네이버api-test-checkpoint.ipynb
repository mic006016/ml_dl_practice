```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request # http통신 라이브러리, requests와 유사
import json # 파이썬에서는 key로 값을 추출하기 위해서는 dict로 변경해주어야한다.

client_id = "UBUeA0P3L9om5VHUFovZ"
client_secret = "rjr9laGWpU"

encText = urllib.parse.quote("가산동 맛집") # utf-8로 인코딩된 텍스트
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
url =  "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=100&start=101"

# http header 만들어주는 부분
# > GET /v1/search/blog.xml?query=%EB%A6%AC%EB%B7%B0&display=10&start=1&sort=sim HTTP/1.1
# > Host: openapi.naver.com
# > User-Agent: curl/7.49.1
# > Accept: */*
# > X-Naver-Client-Id: {애플리케이션 등록 시 발급받은 클라이언트 아이디 값}
# > X-Naver-Client-Secret: {애플리케이션 등록 시 발급받은 클라이언트 시크릿 값}
# >

request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

response = urllib.request.urlopen(request) # 네이버 서버에 http요청
# http응답이 response변수로 쏙 들어감.

# http는 header+body가 있음.
# 응답용 http의 header에는 code + message + http1.1
# 응답용 http의 body에는 전달데이터(json)
rescode = response.getcode() # header에는 code추출(200)
if(rescode==200):
    response_body = response.read() # body에는 전달데이터(json) 추출
    json_data = json.loads(response_body) # json --> dict로 변경
    # print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode) # 200이 아닌 경우 해당 코드 프린트
```

```python
total_list = []
encText = urllib.parse.quote("가산동 맛집")
for i in range(10): # 1~10

    # http 연결
    url =  "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=100&start=" + str(i * 100 + 1)

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    
    response = urllib.request.urlopen(request)
    rescode = response.getcode() 
    if(rescode==200):
        response_body = response.read()
        json_data = json.loads(response_body)
    else:
        print("Error Code:" + rescode)

    # items리스트 가지고 오자.
    items = json_data['items']
        
    # for문 돌려서 items에 있는거 다 배주자.
    for one in items:
        title = one['title']
        link = one['link']
        description = one['description']
        bloggername = one['bloggername']
        bloggerlink = one['bloggerlink']
        postdate = one['postdate']
        
        one_list = [title, link, description, bloggername, bloggerlink, postdate]
        
    # 전체를 모으자.
        total_list.append(one_list)

len(total_list), total_list[999]
```

```python
total_list = [] 
encText = urllib.parse.quote("가산동 맛집")
for i in range(1, 1001, 100): # 1~1000, 100씩 점프

    # s = (i - 1) * 100 + 1
    # i = 1 --> (1 - 1) * 100 + 1 --> 1
    # i = 2 --> (2 - 1) * 100 + 1 --> 101
    # i = 3 --> (3 - 1) * 100 + 1 --> 201
    
    # http 연결
    url =  "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=100&start=" + str(i)

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    
    response = urllib.request.urlopen(request)
    rescode = response.getcode() 
    if(rescode==200):
        response_body = response.read()
        json_data = json.loads(response_body)
    else:
        print("Error Code:" + rescode)

    # items리스트 가지고 오자.
    items = json_data['items']
        
    # for문 돌려서 items에 있는거 다 배주자.
    for one in items:
        title = one['title']
        link = one['link']
        description = one['description']
        bloggername = one['bloggername']
        bloggerlink = one['bloggerlink']
        postdate = one['postdate']
        
        one_list = [title, link, description, bloggername, bloggerlink, postdate]
        
    # 전체를 모으자.
        total_list.append(one_list)

len(total_list), total_list[999]
```

```python
df = pd.DataFrame(total_list,
                  columns=['title', 'link', 'description', 'bloggername', 'bloggerlink', 'postdate']
                 )
```

```python
df.to_csv("가산동맛집.csv", index=False, encoding="UTF-8")
```
