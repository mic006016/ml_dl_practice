```python
import requests

# 1. 기본 URL
url = "http://api.koreafilm.or.kr/openapi-data2/wisenut/search_api/search_json2.jsp"

# 2. 요청 파라미터 (쿼리스트링)
params = {
    "collection": "kmdb_new2",
    "nation": "대한민국",
    "ServiceKey": "9PPC39S328VN8Y6AL120",  # 발급받은 키
    "genre": "범죄",
    "releaseDts": "20250101",
    "listCount": 50
}

# 3. GET 요청 보내기
response = requests.get(url, params=params)

# 4. 결과 출력
print("Status:", response.status_code)
print("Headers:", dict(response.headers))
print("Body:", response.text) # ---> json-viewer로 보기 위해서 json형태의 str을 프린트해봄.
# 파이썬으로 꺼내기 위해서는 dict가 되어야 한다.
```

```python
# http body에 들어있는 json형태의 str을 {키:값}형태의 dict로 변경함.
# 키로 추출하기 위함.
# json viewer 보고 하면 됨.
data = response.json()
# data
```

```python
movies = data["Data"][0]["Result"]
movies[0]
```
