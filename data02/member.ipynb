```python
# member, orders, total
# apple   100     1000
# ice     200     2000
# song    300     3000

# 위 데이터를 데이터 프레임(member_df)에 넣고 출력
# 1.member컬럼 추출후 출력
# 2.member, orders컬럼 추출후 출력
# 3.첫번째 행 추출후 출력
# 4.세번째 행 추출후 출력
# 5.첫번째부터 두번째 행 추출후 출력
# 6.orders컬럼을 추출(orders값이 150이상인 조건)
# 7.total컬럼에 모두 100을 더해 df에 반영, member_df출력
# 8.orders과 total컬럼의 값을 곱해서 result 컬럼에 넣은 후 출력
# 9.song의 값을 winner로 변경후 출력
# 10.인덱스를 member로 변경후 winner라는 인덱스값으로 행 추출 후 출력
```

```python
import pandas as pd
```

```python
member_df = pd.DataFrame({
    'member': ['apple', 'ice', 'song'],
    'orders': [100, 200, 300],
    'total': [1000, 2000, 3000]
})

member_df
```

```python
member_df['member']
```

```python
member_df[['member', 'orders']]
```

```python
member_df['orders'][member_df['orders'] >= 150]
```

```python
member_df['total'] = member_df['total'] + 100
member_df
```

```python
member_df['result'] = member_df['orders'] * member_df['total']
member_df
```

```python
member_df.iloc[2,0] = 'winner'
member_df
```

```python
member_df.set_index('member', inplace=True)
```
