```python
import pandas as pd ## df
import matplotlib.pyplot as plt ## 시각화
import numpy as np ## 수학적인 처리 함수
import seaborn as sns ## 다양한 데이터셋과 시각화 함수
```

```python
# 1. df로 읽어들이고
df = pd.read_csv("scientists.csv", encoding="UTF-8")
df
```

```python
# 2. df의 전체 데이터셋 타입을 확인
df.dtypes
```

```python
# 3. df의 컬럼리스트를 확인
df.columns
```

```python
# 4. df의 값을 확인
df.values
```

```python
# 5. df의 맨 마지막에 price항목을 추가(값은 임의대로 넣음. int)
# df['price'] = 10
# df['price'] = range(1, 9) # row 개수와 일치해야함. range(1, 9) ==> [1,2,3,4,5,6,7,8]
df['price'] = np.random.randint(1,100, size=len(df))
df
```

```python
# 6. df의 맨 마지막에 percent항목을 추가(값의 임의대로 넣음. float)
# df['percent'] = 0.5
df['percent'] = np.random.uniform(1,100, size=len(df))
df['percent'] = df['percent'].round(1)
df
```

```python
# 7. df의 맨 마지막에 sale항목을 추가(price * percent)
df['sale'] = df['price'] * df['percent']
df
```

```python
df['sale'].dtype
```

```python
# 10. df의 Name컬럼의 값이 Rosaline인 행 추출
# df[조건][열][행]
df[df["Name"] == "Rosaline Franklin"]
```

```python
# 11. df의 Name컬럼 중 3~5사이의 행 추출
df['Name'][3:6] # 조건 필요없으면 묶는 대괄호[] 필요 x
```
